```python
import numpy as np
```

```python
# Generate a 1000x1000 matrix with random numbers uniformly distributed
matrix = np.random.rand(1000, 1000)

print(matrix)
```

```python
sum_columns = matrix.sum(axis=0)
#print(f"sum of columns : {sum_columns}")
```

```python
mean_value = np.mean(sum_columns)
std_value = np.std(sum_columns)
#print(f"Column Array: {sum_columns}")
#print(f"Mean: {mean_value:.2f}")
#print(f"Standard Deviation: {std_value:.2f}")
```

```python
import matplotlib.pyplot as plt
plt.plot(sum_columns)
plt.show()
```

```python
# removing the 500 from central the fluctution of the distrbution
centered_data = (sum_columns - 500)

#print(centered_data)
```

```python
plt.hist(centered_data, bins=20, density=True)  # Adjust the number of bins as needed
plt.xlabel("x")
plt.ylabel("y")
plt.title("Distribution of Array Values")
plt.legend('array')
plt.show()
```

```python
import matplotlib.pyplot as plt
s = np.random.normal(mean_value, std_value, 1000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(std_value * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mean_value)**2 / (2 * std_value**2) ),
         linewidth=2, color='r')
plt.xlabel("x")
plt.ylabel("y")
plt.legend('gausssion, hist.array')
plt.title("Distribution of Array Values")
plt.show()
```
